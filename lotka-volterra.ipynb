```python
import numpy as np
import matplotlib.pyplot as plt
import ndes.ndes as ndes
import delfi.delfi as delfi
import compression.score.score as score
import distributions.priors as priors
import tensorflow as tf
import getdist
from getdist import plots, MCSamples
import numpy.random as rng
import simulators.lotka_volterra.lotka_volterra as lv
tf.logging.set_verbosity(tf.logging.ERROR)
```

```python
### SIMULATOR ###

# Lotka-Volterra model set up
init = [50, 100]
dt = 0.2
duration = 30
true_params = [0.01, 0.5, 1.0, 0.01]
max_n_steps = 10000
lv_model = lv.LotkaVolterra(init, true_params)
simulator_args = [lv_model, dt, duration]

# Simulator
def simulator(theta, seed, args, batch):
    
    # Pull out model
    model, dt, duration = args
    
    # Set the seed
    np.random.seed(seed)
    
    # Set parameters
    model.params = np.exp(theta)
    
    # Run simulation
    return model.sim_time(dt, duration)

# Simulate mock data
data = simulator(np.log(true_params), 0, simulator_args, 0)
```

```python
### COMPRESSOR ###

# Compression function
def compressor(data, args):
    return lv.calc_summary_stats(data)

# Compress the data
compressed_data = compressor(data, None)
```

```python
### PRIOR ###
prior = priors.TruncatedGaussian(mean = np.log(true_params), C = np.eye(4)*0.01**2, lower=np.ones(4)*-5, upper=np.ones(4)*2)
```

```python
# Create MAFs
NDEs = [ndes.ConditionalMaskedAutoregressiveFlow(n_inputs=4, n_outputs=9, n_hiddens=[50,50], 
                                               n_mades=5, act_fun=tf.tanh, index = 0)]

# Create the DELFI object
DelfiMAF = delfi.Delfi(compressed_data, prior, NDEs, 
                       param_limits = [prior.lower, prior.upper],
                       param_names = ['\\theta_1', '\\theta_2', '\\theta_3', '\\theta_4'], 
                       results_dir = "simulators/lotka_volterra/results/maf_",
                       input_normalization=None)
```

```python
# Initial samples, batch size for population samples, number of populations
n_initial = 250
n_batch = 250
n_populations = 10

# Do the SNL training
DelfiMAF.sequential_training(simulator, compressor, n_initial, n_batch, n_populations, proposal=prior, patience=10, simulator_args=simulator_args, compressor_args=None, safety=100)
```
